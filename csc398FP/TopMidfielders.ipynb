```python
# Top 10 Midfielders

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'UCL_Midfielders.csv'
midfielders_data = pd.read_csv(file_path, encoding='latin1')

# Calculating a performance metric 
midfielders_data['Performance Metric'] = (
    midfielders_data['Assists'] * 5 +                  
    midfielders_data['Key Passes'] * 3 +             
    midfielders_data['Accurate Passes %'] * 2 +       
    midfielders_data['Accurate Long Balls %'] * 1.5 + 
    midfielders_data['Accurate Passes'] * 0.5         
)

# Top 10 midfielders based on the performance metric
top_10_midfielders = midfielders_data.nlargest(10, 'Performance Metric')
# Print the names
print("Top 10 Midfielders:")
print(top_10_midfielders[['Name', 'Performance Metric']])

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_10_midfielders['Name'], top_10_midfielders['Performance Metric'], alpha=0.7, color='gold')
plt.title("Top 10 Midfielders - Performance Metric")
plt.xlabel("Midfielders")
plt.ylabel("Performance Metric")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
