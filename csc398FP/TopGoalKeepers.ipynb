```python
# Top 10 GoalKeeprs 

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'UCL_Goalkeepers.csv'
goalkeepers_data = pd.read_csv(file_path)

# Calculating a performance metric to rank the goalkeepers (e.g Saves + Clean Sheets + Penalties Saved + Average Sofascore)
goalkeepers_data['Performance Metric'] = (
    goalkeepers_data['Saves'] +
    goalkeepers_data['Clean Sheets'] * 10 +  
    goalkeepers_data['Penalties Saved'] * 5 +
    goalkeepers_data['Average Sofascore']
)
# Top 10 goalkeepers based on the performance metric
top_10_goalkeepers = goalkeepers_data.nlargest(10, 'Performance Metric')

# Print the names 
print("Top 10 Goalkeepers:")
print(top_10_goalkeepers[['Name', 'Performance Metric']])

# Create a bar graph 
plt.figure(figsize=(10, 6))
plt.bar(top_10_goalkeepers['Name'], top_10_goalkeepers['Performance Metric'], alpha=0.7)
plt.title("Top 10 Goalkeepers - Performance Metric")
plt.xlabel("Goalkeepers")
plt.ylabel("Performance Metric")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
