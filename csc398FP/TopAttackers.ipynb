```python
# Top 10 Attackers

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'UCL_ATTACKERS.csv'
attackers_data = pd.read_csv(file_path, encoding='latin1')

# Calculating a performance metric for attackers
attackers_data['Performance Metric'] = (
    attackers_data['Goals'] * 6 +                  
    attackers_data['Assists'] * 4 +               
    attackers_data['Expected Goals (xG)'] * 2 +   
    attackers_data['Shots on Target'] * 1.5 -     
    attackers_data['Big Chances Missed'] * 2 +    
    attackers_data['Goal Conversion %']           
)

# Top 10 attackers based on the performance metric
top_10_attackers = attackers_data.nlargest(10, 'Performance Metric')
# Print the names 
print("Top 10 Attackers:")
print(top_10_attackers[['Name', 'Performance Metric']])

# Create a bar graph 
plt.figure(figsize=(10, 6))
plt.bar(top_10_attackers['Name'], top_10_attackers['Performance Metric'], alpha=0.7, color='orange')
plt.title("Top 10 Attackers - Performance Metric")
plt.xlabel("Attackers")
plt.ylabel("Performance Metric")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
